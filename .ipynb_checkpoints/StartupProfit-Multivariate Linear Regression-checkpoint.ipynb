# Example Multivariate Linear Regression
This notebook contains a walkthrough of a simple multivariate linear regression for calculating the startup profit based on r&d spend,administration and market spend


Import required packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import data set

```python
dataset = pd.read_csv('Startupdata.csv')
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,3]
```

Splitting the data into test set and training set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 1/10)
```

Fitting Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
print('Intercept: \n', regressor.intercept_)
print('Coefficients: \n', regressor.coef_)
```

Predicting the test results

```python
Y_pred = regressor.predict(X_test)
print(Y_pred)
```

Visualising the training set results

```python
print ('Predicted Profit: \n', regressor.predict([[34577 ,6754,786]]))
```

```python
a=float(input('What is the R&D spend ?'))
b=float(input('What is the administration spend ?'))
c=float(input('What is the market spend ?'))
useroutput= regressor.predict([[a,b,c]])
print(useroutput)
```

Visualising relationship between R&D Spend and Profit

```python
plt.scatter(dataset['R&D Spend'], dataset['Profit'], color='red')
plt.title('R&D Spend vs Startup Profit', fontsize=14)
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

Visualising relationship between Administration spend and profit

```python
plt.scatter(dataset['Administration'], dataset['Profit'], color='green')
plt.title('Administration vs Startup Profit', fontsize=14)
plt.xlabel('Administration')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

Visualising Relationship between Marketing spend and Profit

```python
plt.scatter(dataset['Marketing Spend'], dataset['Profit'], color='blue')
plt.title('Marketing Spend vs Startup Profit', fontsize=14)
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```
